# Data Frame
### Whenever there is a scenario that compliesis multiple attributes at a time the data structure we follow is called Data Frame.
### Most probably when we have two dimensional data that data strucuture is nothing but the dataframe

```python
# Import the numpy library for array manipulation and mathematical operations
import numpy as np

# Import the pandas library for data analysis and manipulation
import pandas as pd
```

```python
# Define a dictionary with two Series objects
A = {
    "one": pd.Series([1, 2, 3], index=["a", "b", "c"]),
    "two": pd.Series([1, 2, 3, 4], index=["a", "b", "c", "d"]),
}

# Print the dictionary
print(A)
```

```python
# Create a Series object
series = pd.Series([1, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])

# Print the Series
print(series)
```

```python
# Create multiple Series objects
series1 = pd.Series([10, 20, 30, 40, 50], index=['A', 'B', 'C', 'D', 'E'])
series2 = pd.Series([100, 200, 300, 400, 500], index=['A', 'B', 'C', 'D', 'E'])
series3 = pd.Series([1000, 2000, 3000, 4000, 5000], index=['A', 'B', 'C', 'D', 'E'])
series4 = pd.Series([10000, 20000, 30000, 40000, 50000], index=['A', 'B', 'C', 'D', 'E'])
series5 = pd.Series([11, 21, 31, 41, 51], index=['A', 'B', 'C', 'D', 'E'])
series6 = pd.Series([12, 22, 32, 42, 52], index=['A', 'B', 'C', 'D', 'E'])
series7 = pd.Series([13, 23, 33, 43, 53], index=['A', 'B', 'C', 'D', 'E'])
series8 = pd.Series([14, 24, 34, 44, 54], index=['A', 'B', 'C', 'D', 'E'])
series9 = pd.Series([15, 25, 35, 45, 55], index=['A', 'B', 'C', 'D', 'E'])
series10 = pd.Series([16, 26, 36, 46, 56], index=['A', 'B', 'C', 'D', 'E'])

# Print the Series objects
print(series1)
print(series2)
print(series3)
print(series4)
print(series5)
print(series6)
print(series7)
print(series8)
print(series9)
print(series10)

# Create a DataFrame using the Series objects
DataFrame1 = pd.DataFrame({
    "Column 1": series1,
    "Column 2": series2,
    "Column 3": series3,
    "Column 4": series4,
    "Column 5": series5,
    "Column 6": series6,
    "Column 7": series7,
    "Column 8": series8,
    "Column 9": series9,
    "Column 10": series10,
})

DataFrame1
```

```python
# Create Series objects
series1 = pd.Series(["Ram", "Sham", "Raj", "Dan", "Som", "Tom"])
series2 = pd.Series([50, 54, 29, 37, 60, 78])
series3 = pd.Series(["Animation", "BIA", "Data", "Fintech", "Cloud", "Data"])

# Create a DataFrame using the Series objects
DataFrame2 = pd.DataFrame({
    "Name": series1,
    "Age": series2,
    "Branch": series3,
})

DataFrame2
```
