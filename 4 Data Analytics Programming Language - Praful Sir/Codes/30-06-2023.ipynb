# Create a dataframe using dictionary
### Step 1 :- Create list
### Step 2 :- Convert that list series
### Step 3 :- Convert those series in dataframe using pandas
# Conditions :-
### Dataframe must contain 100 rows & 5 columns
### Column name should be :- S_Col_1, S_Column_2, S_Col_3, Sample_Col_4, Sample_Column_5

```python
# Import the packages

import numpy as np
import pandas as pd
```

```python
# Create the list with 100 Numbers

# List 1 for column 1
List1 = np.arange(100)

# List 2 for column 2
List2 = np.random.randn(100)

# List 3 for column 3
List3 = np.random.randn(100)

# List 4 for column 4
List4 = np.random.randn(100)

# List 5 for column 5
List5 = np.random.randn(100)

# Convert all the list into series
Series1 = pd.Series(List1)
Series2 = pd.Series(List2)
Series3 = pd.Series(List3)
Series4 = pd.Series(List4)
Series5 = pd.Series(List5)

# Create the dataframe
Dataframe1 = pd.DataFrame({
    "S_Col_1":Series1,
    "S_Column_2":Series2,
    "S_Col_3":Series3,
    "Sample_Col_4":Series4,
    "Sample_Column_5":Series5
})

Dataframe1
```
